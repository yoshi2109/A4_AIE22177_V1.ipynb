With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import pandas as pd
import numpy as np

# Load the dataset using double backslashes
data = pd.read_csv("/content/loan-train.csv")

# Or use a raw string
# data = pd.read_csv("/content/loan-train.csv")

# Select two classes (e.g., Loan_Status 'Y' and 'N')
class1_data = data[data['Loan_Status'] == 'Y']
class2_data = data[data['Loan_Status'] == 'N']

# Calculate mean (centroid) for each class
mean_class1 = np.mean(class1_data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']], axis=0)
mean_class2 = np.mean(class2_data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']], axis=0)

# Calculate spread (standard deviation) for each class
spread_class1 = np.std(class1_data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']], axis=0)
spread_class2 = np.std(class2_data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']], axis=0)

# Calculate the distance between mean vectors of the two classes
distance_between_classes = np.linalg.norm(mean_class1 - mean_class2)

# Print results
print("Mean (centroid) for Class 1:", mean_class1)
print("Spread (standard deviation) for Class 1:", spread_class1)
print("\nMean (centroid) for Class 2:", mean_class2)
print("Spread (standard deviation) for Class 2:", spread_class2)
print("\nDistance between mean vectors of Class 1 and Class 2:", distance_between_classes)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/content/loan-train.csv")

# Choose two features for Minkowski distance calculation
feature1 = data['ApplicantIncome']
feature2 = data['CoapplicantIncome']

# Calculate Minkowski distances for r from 1 to 10
r_values = range(1, 11)
distances = [np.linalg.norm(feature1 - feature2, ord=r) for r in r_values]

# Plot the distances
plt.figure(figsize=(10, 6))
plt.plot(r_values, distances, marker='o', linestyle='-')
plt.title('Minkowski Distance between ApplicantIncome and CoapplicantIncome')
plt.xlabel('r (Parameter for Minkowski Distance)')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```

```
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
data = pd.read_csv("/content/loan-train.csv")

# Assuming 'Loan_Status' is the target variable (class levels)
X = data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']]
y = data['Loan_Status']

# Map 'Y' and 'N' to binary values (e.g., 1 and 0)
y = y.map({'Y': 1, 'N': 0})

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Now you have X_train, X_test, y_train, and y_test for further modeling

# Print the shapes of the resulting sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

# Assuming you already have X_train, X_test, y_train, and y_test from the previous exercise

# Handle missing values using SimpleImputer (replace NaN with mean in this case)
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Create a kNN classifier with k=3
neigh = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the imputed training set
neigh.fit(X_train_imputed, y_train)

# Make predictions on the imputed test set
y_pred = neigh.predict(X_test_imputed)

# Calculate and print the accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy on the test set:", accuracy)
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

# Load the dataset
data = pd.read_csv("/content/loan-train.csv")

# Assuming 'Loan_Status' is the target variable (class levels)
X = data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']]
y = data['Loan_Status'].map({'Y': 1, 'N': 0})

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Handle missing values using SimpleImputer (replace NaN with mean)
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Vary k from 1 to 11
k_values = range(1, 12)
accuracies = []

for k in k_values:
    # Create a kNN classifier with the current k value
    neigh = KNeighborsClassifier(n_neighbors=k)

    # Train the classifier on the imputed training set
    neigh.fit(X_train_imputed, y_train)

    # Make predictions on the imputed test set
    y_pred = neigh.predict(X_test_imputed)

    # Calculate accuracy on the test set
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot the accuracy for different k values
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.title('Accuracy vs. k for kNN Classifier')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```
from sklearn.metrics import confusion_matrix, classification_report

# Assuming you already have X_train, X_test, y_train, y_test, X_train_imputed, X_test_imputed, and neigh from previous exercises

# Predictions on training set
y_train_pred = neigh.predict(X_train_imputed)

# Predictions on test set
y_test_pred = neigh.predict(X_test_imputed)

# Confusion matrix and classification report for training set
conf_matrix_train = confusion_matrix(y_train, y_train_pred)
report_train = classification_report(y_train, y_train_pred)

# Confusion matrix and classification report for test set
conf_matrix_test = confusion_matrix(y_test, y_test_pred)
report_test = classification_report(y_test, y_test_pred)

# Print confusion matrices and classification reports
print("Confusion Matrix for Training Set:")
print(conf_matrix_train)

print("\nClassification Report for Training Set:")
print(report_train)

print("\nConfusion Matrix for Test Set:")
print(conf_matrix_test)

print("\nClassification Report for Test Set:")
print(report_test)
```
